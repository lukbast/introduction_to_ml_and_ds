## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for ploting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # width and height

# 3. Plot data

ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
# 5. Save & show (you save the whole figure)
fig.savefig("img/sample-plot.png")
```

## 2 options for subplots

```python
# option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
ax1.set(title="ax1");
ax1.plot(x, x/2)
ax2.set(title="ax2");
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.set(title="ax3");
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.set(title="ax4");
ax4.hist(np.random.random(10));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,8))
# plot to each different index
ax[0, 0].set(title="ax[0,0]");
ax[0, 0].plot(x, x/2)
ax[0, 1].set(title="ax[0,1]");
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].set(title="ax[1,0]");
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].set(title="ax[1,1]");
ax[1, 1].hist(np.random.random(10));
```

### Which one should you use? (pyplot vs matplotlib OO Method)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method 

```python
heart_diesease.head()
```

```python
over_50 = heart_diesease[heart_diesease["age"] > 50]
```

```python
len(heart_diesease)
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# combination of plot() and OO method
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
ax.set_xlim([45,80])
```

```python
# OO from scratch
fig, ax = plt.subplots(figsize=(10,6))
# plot the data
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"])

# customize
ax.set_xlim(45,80)
ax.set(title="Cholestelor in group of patients over 50",
       xlabel="Age", ylabel="Cholestelor")
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(), ls="--");
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)

# add data to axes
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],
                     c=over_50["target"])

scatter_thalach= ax1.scatter(x=over_50["age"], y=over_50["thalach"],
                     c=over_50["target"])

# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age", ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(),
           ls="--");

# customize ax1
ax1.set(title="Maximum heart rate achieved and Heart Diesease", xlabel="Age", ylabel="Maximum heart rate achieved");
ax1.legend(*scatter_thalach.legend_elements(), title="Target");
ax1.axhline(over_50["thalach"].mean());

fig.suptitle("Heart Diesease Analisys",fontsize=24, weight="bold")
```
